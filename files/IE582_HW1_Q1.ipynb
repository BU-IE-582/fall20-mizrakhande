```r
data1= read.csv("E0_2020_2021.csv",header = TRUE)
data2= read.csv("E0_2019_2020.csv",header = TRUE)
data3= read.csv("E0_2018_2019.csv",header = TRUE)
data1=data1[,c(2,4:8)]
data2=data2[,c(2,4:8)]
data3=data3[,c(2:7)]
all_data=rbind(data1,data2,data3)
```

I combined English Premier League data of Season
2020/2021, Season 2019/2020 and Season 2018/2019.

```r
home_goals=all_data$FTHG
away_goals=all_data$FTAG
difference=all_data$FTHG-all_data$FTAG
```

Then, I took full time home goals and full time away goals columns. 

```r
hist(home_goals,
     xlab="Home Goals",ylab = " Number of Games",ylim = c(0,500),right=FALSE)
hist(away_goals,
     xlab="Away Goals",ylab = " Number of Games",ylim = c(0,600),right=FALSE)
hist(difference,
     xlab="Home Goals-Away Goals",ylab = " Number of Games",ylim = c(0,500),right=FALSE)
difference_density=density(difference)
plot(difference_density)
polygon(difference_density, col="red", border="blue")
```

```r
expectedvaluehome=c()
samplemeans=mean(home_goals)
for (j in 0:max(home_goals)){
    expectedvaluehome[j+1]=dpois(j, samplemeans)*length(home_goals)
  }
hist(home_goals,xlab="Home Goals",ylab = " Number of Games",ylim = c(0,300),right=FALSE)
par(new=TRUE)
plot(c(0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5,8.5), expectedvaluehome, type= "b",xaxt="n", yaxt="n", ann=FALSE,col="red",xlim=c(0,8))
```

The distributions of home and away goals look like Poisson distribution. To prove that mean of home_goals is calculated and it is used as  of a Poisson distribution. By using dpois(#ofGoals, ) command the probability of getting that amount of goal is calculated. Subsequently, this probability is multiplied by number of observations (number of total games) to get expected value. Then, it is plotted to the home_goals histogram.

```r
expectedvaluehome=c()
samplemeans=mean(away_goals)
for (j in 0:max(away_goals)){
    expectedvaluehome[j+1]=dpois(j, samplemeans)*length(away_goals)
  }
hist(away_goals,xlab="Away Goals",ylab = " Number of Games",ylim = c(0,300),right=FALSE)
par(new=TRUE)
plot(c(0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5,8.5,9.5), expectedvaluehome, type= "b",xaxt="n", yaxt="n", ann=FALSE,col="red",xlim=c(0,9))
```

The same procedure is applied for away goals.
